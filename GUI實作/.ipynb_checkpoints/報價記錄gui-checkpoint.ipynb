```python
main = Tk()
main.title("測試區")
main.geometry("400x600")

def open_file():
    root = Tk()
    root.withdraw()
    file_path = filedialog.askopenfilename()
    messagebox.showinfo(title='開啟檔案', message = file_path)
    return file_path

menubar = Menu(main)
filemenu = Menu(menubar, tearoff=0)
filemenu.add_command(label="New", command=donothing)
filemenu.add_command(label="Open", command=open_file)
filemenu.add_command(label="Save", command=donothing)
filemenu.add_command(label="Save as...", command=donothing)
filemenu.add_command(label="Close", command=donothing)
filemenu.add_command(label="Exit", command=main.quit)
menubar.add_cascade(label="File", menu=filemenu)
editmenu = Menu(menubar, tearoff=0)
editmenu.add_command(label="Undo", command=donothing)

editmenu.add_separator()

editmenu.add_command(label="Cut", command=donothing)
editmenu.add_command(label="Copy", command=donothing)
editmenu.add_command(label="Paste", command=donothing)
editmenu.add_command(label="Delete", command=donothing)
editmenu.add_command(label="Select All", command=donothing)

menubar.add_cascade(label="Edit", menu=editmenu)
helpmenu = Menu(menubar, tearoff=0)
helpmenu.add_command(label="Help Index", command=donothing)
helpmenu.add_command(label="About...", command=donothing)
menubar.add_cascade(label="Help", menu=helpmenu)

label = Label(main, text='Test')
label.grid(row=0, column=0)

region = Label(main, text='Conutry')
region.grid(row=1, column=0)

region_list = [
    "India",
    "Russia",
    "OEM/ODM"
]
variable = StringVar(main)
variable.set("")
country_list = OptionMenu(main, variable,*region_list)
country_list.grid(row=1, column=1)

main.config(menu=menubar)
main.mainloop()
```

## 正式區

```python
root = Tk()
root.title("報價記錄維護")
root.geometry('400x600')

region_list = [
    "India",
    "Russia",
    "OEM/ODM"
]

l1 = Label(root, text="區域:")
l1.pack()  #這裡的side可以賦值為LEFT  RTGHT TOP  BOTTOM
# xls_text = StringVar()
# xls = Entry(root, textvariable = xls_text)
# xls.OptionMenu(root, *region_list)
# xls_text.set("")
# xls.pack(side = "left")
variable = StringVar()
variable.set("")
w = OptionMenu(root, variable, *region_list)
w.pack()


l2 = Label(root, text="代理商:")
l2.pack()  #這裡的side可以賦值為LEFT  RTGHT TOP  BOTTOM
sheet_text = StringVar()
sheet = Entry(root, textvariable = sheet_text)
sheet_text.set("")
sheet.pack()


l3 = Label(root, text="終端客戶:")
l3.pack()  #這裡的side可以賦值為LEFT  RTGHT TOP  BOTTOM
loop_text = StringVar()
loop = Entry(root, textvariable = loop_text)
loop_text.set("")
loop.pack()

l4 = Label(root, text="應用:")
l4.pack()  #這裡的side可以賦值為LEFT  RTGHT TOP  BOTTOM
sleep_text = StringVar()
sleep = Entry(root, textvariable = sleep_text)
sleep_text.set("")
sleep.pack()

l5 = Label(root, text="供應商:")
l5.pack()  #這裡的side可以賦值為LEFT  RTGHT TOP  BOTTOM
supplier_text = StringVar()
supplier = Entry(root, textvariable = supplier_text)
supplier_text.set("")
supplier.pack()

l6 = Label(root, text="供應商PN:")
l6.pack()
supplier_pn_text = StringVar()
supplier_pn = Entry(root, textvariable = supplier_pn_text)
supplier_pn_text.set("")
supplier_pn.pack()

l7 = Label(root, text="目標價格:")
l7.pack()
tar_price_text = StringVar()
tar_price = Entry(root, textvariable = tar_price_text)
tar_price_text.set("")
tar_price.pack()

l8 = Label(root, text="INPAQ報價:")
l8.pack()
inpaq_price_text = StringVar()
inpaq_price = Entry(root, textvariable = inpaq_price_text)
inpaq_price_text.set("")
inpaq_price.pack()

unit_list = [
    "KPCS",
    "PCS"
]

l9 = Label(root, text="單位:")
l9.pack()
unit_var = StringVar(root)
unit_var.set("")
unit = OptionMenu(root, unit_var, *unit_list)
unit.pack()

cur_list = [
    "USD", "NTD", "CNY", "EUR"
]
cur = Label(root, text="幣別:")
cur.pack()
cur_var = StringVar(root)
cur_var.set("")
currency = OptionMenu(root, cur_var, *cur_list)
currency.pack()

l10 = Label(root, text="成本:")
l10.pack()
cost_text = StringVar()
cost = Entry(root, textvariable = cost_text)
cost_text.set("")
cost.pack()


def on_click():
    x = variable.get()
    s = sheet.get()
    l = loop.get()
    sl = sleep.get()
    sup = supplier.get()
    sup_pn = supplier.get()
    tar_pri = tar_price.get()
    inpaq_pri = inpaq_price.get()
    unit = unit_var.get()
    cos = cost.get()
    currency = cur_var.get()
    today = datetime.today().strftime("%Y-%m-%d")
    
    string = str("日期:%s 區域:%s 代理商:%s 終端客戶:%s 應用:%s 供應商:%s 供應商PN:%s 目標價格:%s INPAQ報價:%s 單位:%s 幣別:%s 成本:%s" %(today, x, s, l, sl, sup, sup_pn, tar_pri, inpaq_pri, unit, currency, cos))
    print("日期:%s 區域:%s 代理商:%s 終端客戶:%s 應用:%s 供應商:%s 供應商PN:%s 目標價格:%s INPAQ報價:%s 單位:%s 幣別:%s 成本:%s" %(today, x, s, l, sl, sup, sup_pn, tar_pri, inpaq_pri, unit, currency, cos))
    messagebox.showinfo(title='輸入成功', message = string)
    df = pd.read_excel(r"D:\報價\報價記錄\2021報價記錄維護表.xlsx")
    df = df.append({"日期": today, "區域": x, "代理商": s, "終端客戶": l, "應用": sl, "供應商": sup, "供應商PN": sup_pn, 
              "目標價格": tar_pri, "INPAQ報價": inpaq_pri, "單位": unit, "幣別": currency, "成本": cos}, ignore_index=True)
    df.to_excel(r"D:\報價\報價記錄\2021報價記錄維護表.xlsx", index=False)
    
    
def close_win():
    root.destroy()
#     root.quit()
    
def clean_text():
#     w.delete(0, END)
    sheet.delete(0, END)
    loop.delete(0, END)
    sleep.delete(0, END)
    supplier.delete(0, END)
    supplier_pn.delete(0, END)
    tar_price.delete(0, END)
    inpaq_price.delete(0, END)
#     unit.delete(0, END)
    cost.delete(0, END)
    
def quotation_record():
    quote = Tk()
    quote.title("報價查詢（請輸入品名）")
    quote.geometry('400x600')
    greeting = Label(quote, text="敬請期待~~~\n產品功能開發中")
    greeting.pack()
    

def quotation():
    var1 = cost_num.get()
    var2 = eau.get()
    if int(var2) >= 10000:
        quote = round(float(var1) / 0.8, 2)
    elif 5000 <= int(var2) < 10000:
        quote = round(float(var1) / 0.75, 2)
    else:
        quote = round(float(var1) / 0.7, 2)
    result.insert('1.0', quote)
        
    
def quotation_calculation():
    calculator = Tk()
    calculator.title("報價計算")
    calculator.geometry('400x600')
    global eau, cost_num, result
    instruction = Label(calculator, text="請輸入產品成本").grid(row=0, column=0, sticky=W, pady=2)
    cost_num_text = DoubleVar()
    cost_num = Entry(calculator, textvariable = cost_num_text).grid(row=0, column=1, sticky=W, pady=2)
    volume_ins = Label(calculator, text="請輸入EAU").grid(row=1, column=0, sticky=W, pady=2)
    eau_text = IntVar()
    eau = Entry(calculator, textvariable = eau_text).grid(row=1, column=1, sticky=W, pady=2)
    calculate = Button(calculator, text = "計算", command=quotation, width=10, height=2).grid(row=3, column=1, sticky=E, pady=2)
    result = Text(calculator, height=1, textvariable=cost_num.get())
    result.grid(row=3, column=0, sticky=W, pady=2, rowspan=1)


    
    
    
Button(root, text="輸入", command = on_click).pack()
Button(root, text="下一筆", command= clean_text).pack()
Button(root, text="結束程式", command = close_win).pack()
Button(root, text="報價查詢", command = quotation_record).pack()
Button(root, text="報價計算", command = quotation_calculation).pack()
    
root.mainloop()
```
