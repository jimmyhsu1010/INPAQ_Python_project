```python
import pandas as pd
pd.set_option("display.max_rows", 8)
pd.set_option("display.max_columns", None)
pd.set_option("display.float_format", lambda x: "%.0f" % x)
import tkinter as tk
from tkinter import filedialog
```

```python
root = tk.Tk()
root.withdraw()
file_path = filedialog.askopenfilename()
```

```python
df = pd.read_excel(file_path, None)
```

```python
sheet_dict = {}
for k, v in enumerate(list(df.keys())):
    print(k+1, v)
    sheet_dict.setdefault(k+1, v)

sheet_name = eval(input("請選擇要開啟的sheet："))
df = df[sheet_dict[sheet_name]]
```

```python
df
```

```python
filtered = df[(df["狀態"].str.contains("出")) & (df["BG"] == "RF") & (df["預交年份"] == 2021) & (df["預交月份"].isin(["January", "February", "March"]))]
```

```python
filtered["預交月份"] = pd.Categorical(filtered["預交月份"], ordered=True, categories=["January", "February", "March"])
```

```python
filtered["BG"] = filtered.apply(lambda x: "BU2" if x["品名"][0:4] == "RFDP" else "BU1", axis=1)
```

```python
pivot = pd.pivot_table(filtered, index=["負責業務", "BG", "Group", "品名", "單價"], columns="預交月份", values=["數量", "本國幣別NTD"], aggfunc="sum")
```

```python
sum_table = pd.pivot_table(filtered, index=["負責業務", "BG"], columns="預交月份", values=["本國幣別NTD"], aggfunc="sum")
```
