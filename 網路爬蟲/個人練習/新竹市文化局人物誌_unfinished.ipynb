```python
headers = {'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 11_2_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.114 Safari/537.36', 'content-type': 'text/plain'}
url = 'https://culture.hccg.gov.tw/ch/home.jsp?id=285&parentpath=0,3&mcustomize=&serno=null&page='
directory = 'https://culture.hccg.gov.tw/ch/'
```

```python
def get_links(url, headers, directory):
    links = []
    for i in range(1, 28):
        url = url + str(i)
        response = requests.get(url, headers=headers)
#         print(response)
        soup = BeautifulSoup(response.content, 'html.parser')
        all_divs = soup.find_all('div', class_='css_tr list_list')
        print(all_divs)
        for i in all_divs:
            try:
                a_label = i.find('a').get('href')
            except:
                pass
            else:
                article_link = directory + a_label
                print(article_link)
                links.append(article_link)
        time.sleep(randint(3,5))
    return links
```
